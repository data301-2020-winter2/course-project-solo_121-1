```python
import pandas as pd
MedDataSet = pd.read_csv("Medical_Cost.csv")
MedDataSet
```

```python
def load_and_process(url_or_path_to_csv_file):
    df = (
        pd.read_csv(url_or_path_to_csv_file)
        .rename(columns ={"bmi": "BMI"})
        .dropna()
     ) 
    
    return df
load_and_process("Medical_Cost.csv")
```

```python
#1
import seaborn as sns
sns.set(rc={'figure.figsize':(11.7,8.27)})
sns.set_theme(style="darkgrid")
G1 = sns.countplot(x="age",hue="smoker", data=MedDataSet)

#In this graph you can see that at young ages smokers are very high but with the rest the smoking numbers is around an average for them all.
```
